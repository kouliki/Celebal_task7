###Handling Missing Values

```python
import pandas as pd

# Select only numeric columns before filling missing values
numeric_columns = train.select_dtypes(include=['number']).columns
train[numeric_columns] = train[numeric_columns].fillna(train[numeric_columns].median(), inplace=False)
```

```python
import pandas as pd

# Select only numeric columns before filling missing values
numeric_columns = test.select_dtypes(include=['number']).columns
test[numeric_columns] = test[numeric_columns].fillna(test[numeric_columns].median(), inplace=False)
# inplace=False to create a copy and avoid modifying original DataFrame
```
